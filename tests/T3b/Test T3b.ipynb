```python
import matplotlib
import sympy
from sympy import init_printing, cos, sin
init_printing()
```

```python
pi = sympy.symbols('pi')
x = sympy.symbols('x')
y = sympy.symbols('y')
z = sympy.symbols('z')

lx = sympy.symbols('l_x')
ly = sympy.symbols('l_y')
lz = sympy.symbols('l_z')

X = (2 * pi) * (x / lx)
Y = (2 * pi) * (y / ly)
Z = (2 * pi) * (z / lz)
```

```python
# Define velocity field
u = (lx / (2 * pi)) * sin(X) * cos(Y) * cos(Z)
v = (ly / (2 * pi)) * cos(X) * sin(Y) * cos(Z)
w = -2 * (lz / (2 * pi)) * cos(X) * cos(Y) * sin(Z)

# Check it is divergence free
divu = (sympy.diff(u, x) + sympy.diff(v, y) + sympy.diff(w, z))
divu.simplify()
```

```python
# Define density field
rho0 = sympy.symbols('rho_0')
rho = rho0 + sin(X) * sin(Y) * sin(Z)

# Check u dot grad rho = 0
drdx = sympy.diff(rho, x)
drdy = sympy.diff(rho, y)
drdz = sympy.diff(rho, z)
ugr = u * drdx
vgr = v * drdy
wgr = w * drdz
ugr + vgr + wgr
```

```python
# Compute MMS source term

Re = sympy.symbols('Re')
Pr = sympy.symbols('Pr')
p0 = sympy.symbols('p_0')

T = p0 / rho
dTdx = sympy.diff(T, x)
dTdy = sympy.diff(T, y)
dTdz = sympy.diff(T, z)

d2Tdx2 = sympy.diff(dTdx, x)
d2Tdy2 = sympy.diff(dTdy, y)
d2Tdz2 = sympy.diff(dTdz, z)
T.simplify()
```

```python
SrhoX = d2Tdx2.simplify()
SrhoX
```

```python
SrhoY = d2Tdy2.simplify()
SrhoY
```

```python
SrhoZ = d2Tdz2.simplify()
SrhoZ
```

```python
# Compute divu

divu = ((1 / (Re * Pr * T)) * (d2Tdx2 + d2Tdy2 + d2Tdz2)) / rho
divu.simplify()
```

```python
# Compute density source
S = rho * divu + (ugr + vgr + wgr)
S.simplify()
```

```python
# Compute source terms for momentum

# Advection
#SX = u * sympy.diff(u, x) + v * sympy.diff(u, y) + w * sympy.diff(u,z)
#SX *= rho
#SY = u * sympy.diff(v, x) + v * sympy.diff(v, y) + w * sympy.diff(v,z)
#SY *= rho
#SZ = u * sympy.diff(w, x) + v * sympy.diff(w, y) + w * sympy.diff(w,z)
#SZ *= rho

SX = 0
SX += sympy.diff(rho * u * u, x) / 2
SX += sympy.diff(rho * u * v, y) / 2
SX += sympy.diff(rho * u * w, z) / 2
SX += rho * u * sympy.diff(u, x) / 2
SX += rho * v * sympy.diff(u, y) / 2
SX += rho * w * sympy.diff(u, z) / 2

SY = 0
SY += sympy.diff(rho * v * u, x) / 2
SY += sympy.diff(rho * v * v, y) / 2
SY += sympy.diff(rho * v * w, z) / 2
SY += rho * u * sympy.diff(v, x) / 2
SY += rho * v * sympy.diff(v, y) / 2
SY += rho * w * sympy.diff(v, z) / 2

SZ = 0
SZ += sympy.diff(rho * w * u, x) / 2
SZ += sympy.diff(rho * w * v, y) / 2
SZ += sympy.diff(rho * w * w, z) / 2
SZ += rho * u * sympy.diff(w, x) / 2
SZ += rho * v * sympy.diff(w, y) / 2
SZ += rho * w * sympy.diff(w, z) / 2
```

```python
# XMOM 
SX.simplify()
```

```python
# YMOM 
SY.simplify()
```

```python
# ZMOM 
SZ.simplify()
```

```python
# Viscous stress (Incompressible)
dudx = sympy.diff(u, x)
dudy = sympy.diff(u, y)
dudz = sympy.diff(u, z)
d2udx2 = sympy.diff(dudx, x)
d2udy2 = sympy.diff(dudy, y)
d2udz2 = sympy.diff(dudz, z)
SX = -(1 / Re) * (d2udx2 + d2udy2 + d2udz2)
dvdx = sympy.diff(v, x)
dvdy = sympy.diff(v, y)
dvdz = sympy.diff(v, z)
d2vdx2 = sympy.diff(dvdx, x)
d2vdy2 = sympy.diff(dvdy, y)
d2vdz2 = sympy.diff(dvdz, z)
SY = -(1 / Re) * (d2vdx2 + d2vdy2 + d2vdz2)
dwdx = sympy.diff(w, x)
dwdy = sympy.diff(w, y)
dwdz = sympy.diff(w, z)
d2wdx2 = sympy.diff(dwdx, x)
d2wdy2 = sympy.diff(dwdy, y)
d2wdz2 = sympy.diff(dwdz, z)
SZ = -(1 / Re) * (d2wdx2 + d2wdy2 + d2wdz2)
```

```python
SX.simplify()
```

```python
# Full viscous stress tensor

dvdx = sympy.diff(v, x)
dwdx = sympy.diff(w, x)
d2vdxdy = sympy.diff(dvdx, y)
d2wdxdz = sympy.diff(dwdx, z)
SX -= (1 / Re) * (d2udx2 + d2vdxdy + d2wdxdz)
BX = (1 / Re) * sympy.Rational(2, 3) * sympy.diff(divu, x)

dudy = sympy.diff(u, y)
dwdy = sympy.diff(w, y)
d2udydx = sympy.diff(dudy, x)
d2wdydz = sympy.diff(dwdy, z)
SY -= (1 / Re) * (d2udydx + d2vdy2 + d2wdydz)
BY = (1 / Re) * sympy.Rational(2, 3) * sympy.diff(divu, y)

dudz = sympy.diff(u, z)
dvdz = sympy.diff(v, z)
d2udzdx = sympy.diff(dudz, x)
d2vdzdy = sympy.diff(dvdz, y)
SZ -= (1 / Re) * (d2udzdx + d2vdzdy + d2wdz2)
BZ = (1 / Re) * sympy.Rational(2, 3) * sympy.diff(divu, z)
```
