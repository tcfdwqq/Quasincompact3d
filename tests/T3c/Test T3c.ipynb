# Test T3c

This is based on the pseudo 'Taylor-Green' testcase of `Shunn2012` but without a pressure update.

The solution is given as:

$\phi = \frac{1 + sin(k \pi \widehat{x}) sin(k \pi \widehat{y}) cos(\omega \pi t)}{\left(1 + \frac{\rho_0}{\rho_1} \right)
+ \left(1 - \frac{\rho_0}{\rho_1} \right) sin(k \pi \widehat{x}) sin(k \pi \widehat{y}) cos(\omega \pi t)}$

$\rho = {\left( \frac{\phi}{\rho_1} + \frac{1 - \phi}{\rho_0} \right)}^{-1}$

$u = \frac{\rho_1 - \rho_0}{\rho} \frac{-\omega}{4 k} cos(k \pi \widehat{x}) sin(k \pi \widehat{y}) sin(\omega \pi t)$

$v = \frac{\rho_1 - \rho_0}{\rho} \frac{-\omega}{4 k} sin(k \pi \widehat{x}) cos(k \pi \widehat{y}) sin(\omega \pi t)$

$p = 0$

$\widehat{x} = x - u_F t$

$\widehat{y} = y - v_F t$

```python
# Import stuff
import sympy
from sympy import init_printing, cos, sin
init_printing(use_latex=True)
```

```python
# Setup variables

x = sympy.Symbol('x')
y = sympy.Symbol('y')
t = sympy.Symbol('t')

uf = 0#sympy.Symbol('u_F')
vf = 0#sympy.Symbol('v_F')

k = sympy.Symbol('k')
w = sympy.Symbol('omega')
PI = sympy.pi

xhat = x - uf * t
yhat = y - vf * t

rho0 = sympy.Symbol('rho_0')
rho1 = sympy.Symbol('rho_1')

press0 = sympy.Symbol('p_0')

Re = sympy.Symbol('Re')
Pr = sympy.Symbol('Pr')

phi = 1 + sympy.sin(PI * k * xhat) * sympy.sin(PI * k * yhat) \
    * sympy.cos(PI * w * t)
phi /= ((1 + rho0 / rho1) + (1 - rho0 / rho1) \
    * sympy.sin(PI * k * xhat) * sympy.sin(PI * k * yhat) \
    * sympy.cos(PI * w * t))
rho = 1 / (phi / rho1 + (1 - phi) / rho0)
T = press0 / rho

u = ((rho1 - rho0) / rho) * (-w / (4 * k)) \
    * sympy.cos(PI * k * xhat) * sympy.sin(PI * k * yhat) \
    * sympy.sin(PI * w * t)
v = ((rho1 - rho0) / rho) * (-w / (4 * k)) \
    * sympy.sin(PI * k * xhat) * sympy.cos(PI * k * yhat) \
    * sympy.sin(PI * w * t)

rho.simplify()
```

```python
# Check continuity
cont = sympy.diff(rho, t).simplify() \
    + sympy.diff(rho * u, x).simplify() \
    + sympy.diff(rho * v, y).simplify()
cont.simplify()
```

```python
# Compute Laplacian of Temperature

dTdx = sympy.diff(T, x)
dTdy = sympy.diff(T, y)

d2Tdx2 = sympy.diff(dTdx, x)
d2Tdy2 = sympy.diff(dTdy, y)

TLapl = d2Tdx2 + d2Tdy2
```

```python
d2Tdx2.simplify()
```

```python
d2Tdy2.simplify()
```

```python
# Compute divu

divu = (1 / (Re * Pr * T)) * TLapl / rho
divu.simplify()
```

```python
# Compute density source term

drhodt = sympy.diff(rho, t)
drhodt.simplify()
```

```python
udrhodx = u * sympy.diff(rho, x)
#udrhodx.simplify()
sympy.diff(rho, x).simplify()
```

```python
vdrhody = v * sympy.diff(rho, y)
#vdrhody.simplify()
sympy.diff(rho, y).simplify()
```

```python
rhodivu = rho * divu
rhodivu.simplify()
```

```python
# Compute velocity derivatives

dudx = sympy.diff(u, x)
dudy = sympy.diff(u, y)
d2udx2 = sympy.diff(dudx, x)
d2udy2 = sympy.diff(dudy, y)
d2udxdy = sympy.diff(dudx, y)

dvdx = sympy.diff(v, x)
dvdy = sympy.diff(v, y)
d2vdx2 = sympy.diff(dvdx, x)
d2vdy2 = sympy.diff(dvdy, y)
d2vdxdy = sympy.diff(dvdx, y)
```
