```python
import matplotlib
import sympy
from sympy import init_printing, cos, sin
init_printing()
```

```python
x = sympy.symbols('x')
y = sympy.symbols('y')
z = sympy.symbols('z')

lx = sympy.symbols('l_x')
ly = sympy.symbols('l_y')
lz = sympy.symbols('l_z')

X = (2 * sympy.pi) * (x / lx)
Y = (2 * sympy.pi) * (y / ly)
Z = (2 * sympy.pi) * (z / lz)
```

```python
# Define velocity field
u = (lx / (2 * sympy.pi)) * sin(X) * cos(Y) * cos(Z)
v = (ly / (2 * sympy.pi)) * cos(X) * sin(Y) * cos(Z)
w = -2 * (lz / (2 * sympy.pi)) * cos(X) * cos(Y) * sin(Z)

# Check it is divergence free
divu = (sympy.diff(u, x) + sympy.diff(v, y) + sympy.diff(w, z))
divu.simplify()
```

```python
# Define density field
rho0 = sympy.symbols('rho_0')
rho = rho0 + sin(X) * sin(Y) * sin(Z)

# Check u dot grad rho = 0
drdx = sympy.diff(rho, x)
drdy = sympy.diff(rho, y)
drdz = sympy.diff(rho, z)
ugr = u * drdx
vgr = v * drdy
wgr = w * drdz
ugr + vgr + wgr
```

```python
# Compute MMS source term

Re = sympy.symbols('Re')
Pr = sympy.symbols('Pr')

irho = 1 / rho
dirdx = sympy.diff(irho, x)
dirdy = sympy.diff(irho, y)
dirdz = sympy.diff(irho, z)

d2irdx2 = sympy.diff(dirdx, x)
d2irdy2 = sympy.diff(dirdy, y)
d2irdz2 = sympy.diff(dirdz, z)
```

```python
SrhoX = d2irdx2.simplify()
SrhoX
```

```python
SrhoY = d2irdy2.simplify()
SrhoY
```

```python
SrhoZ = d2irdz2.simplify()
SrhoZ
```

```python
S = (rho / (Re * Pr)) * (SrhoX + SrhoY + SrhoZ)
S.simplify()
```

```python
# Compute dt to give maximum CFL = 0.1 on 128^3 grid
# Also time to run through domain 10 times

l = 1.0
n = 128
PI = 3.14
CFL = 0.1

U = 2 * (l / (2 * PI))
dx = l / n
dt = CFL * dx / U

u = 1 * (l / (2 * PI)) # Not really minimum velocity (that would be 0)
t_thru = l / u
t_end = 10 * t_thru
nsteps = t_end / dt

print "dt = ", dt
print "t_end = ", t_end, " nsteps = ", nsteps
```

```python
# Compute "true" CFL
dt = 0.002
U * dt / dx
```

```python
# Compute source terms for momentum

# Advection
SX = u * sympy.diff(u, x) + v * sympy.diff(u, y) + w * sympy.diff(u,z)
SX *= rho
SY = u * sympy.diff(v, x) + v * sympy.diff(v, y) + w * sympy.diff(v,z)
SY *= rho
SZ = u * sympy.diff(w, x) + v * sympy.diff(w, y) + w * sympy.diff(w,z)
SZ *= rho

# Viscous stress (Incompressible)
dudx = sympy.diff(u, x)
dudy = sympy.diff(u, y)
dudz = sympy.diff(u, z)
d2udx2 = sympy.diff(dudx, x)
d2udy2 = sympy.diff(dudy, y)
d2udz2 = sympy.diff(dudz, z)
SX -= (1 / Re) * (d2udx2 + d2udy2 + d2udz2)
dvdx = sympy.diff(v, x)
dvdy = sympy.diff(v, y)
dvdz = sympy.diff(v, z)
d2vdx2 = sympy.diff(dvdx, x)
d2vdy2 = sympy.diff(dvdy, y)
d2vdz2 = sympy.diff(dvdz, z)
SY -= (1 / Re) * (d2vdx2 + d2vdy2 + d2vdz2)
dwdx = sympy.diff(w, x)
dwdy = sympy.diff(w, y)
dwdz = sympy.diff(w, z)
d2wdx2 = sympy.diff(dwdx, x)
d2wdy2 = sympy.diff(dwdy, y)
d2wdz2 = sympy.diff(dwdz, z)
SZ -= (1 / Re) * (d2wdx2 + d2wdy2 + d2wdz2)
```

```python
# XMOM (incompressible viscous stress)
SX.simplify()
```

```python
# YMOM (incompressible viscous stress)
SY.simplify()
```

```python
# ZMOM (incompressible viscous stress)
SZ.simplify()
```

```python
# Full shear stress tensor

dvdx = sympy.diff(v, x)
dwdx = sympy.diff(w, x)
d2vdxdy = sympy.diff(dvdx, y)
d2wdxdz = sympy.diff(dwdx, z)
SX -= (1 / Re) * (d2udx2 + d2vdxdy + d2wdxdz)
SX += (1 / Re) * (2 / 3) * sympy.diff(divu, x)

dudy = sympy.diff(u, y)
dwdy = sympy.diff(w, y)
d2udydx = sympy.diff(dudy, x)
d2wdydz = sympy.diff(dwdy, z)
SY -= (1 / Re) * (d2udydx + d2vdy2 + d2wdydz)
SY += (1 / Re) * (2 / 3) * sympy.diff(divu, y)

dudz = sympy.diff(u, z)
dvdz = sympy.diff(w, z)
d2udzdx = sympy.diff(dudz, x)
d2vdzdy = sympy.diff(dvdz, y)
SZ -= (1 / Re) * (d2udzdx + d2vdzdy + d2wdz2)
SZ += (1 / Re) * (2 / 3) * sympy.diff(divu, z)
```
